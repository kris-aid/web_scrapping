# 1. ¿Qué es el Web Scraping? 🕵️‍♂️
Antes de tirar nuestra primera línea de código, necesitamos entender qué estamos haciendo realmente cuando "scrapeamos" una web.

## 1.1 Definición de Web Scraping
En términos simples, el Web Scraping es el proceso de automatizar la extracción de datos de un sitio web.

Imagina que necesitas los precios de 500 libros de una tienda en línea para tu investigación. Podrías hacerlo a mano: abrir cada página, copiar el precio y pegarlo en un Excel. Eso te tomaría horas (y terminarías odiando la investigación). El **scraping** consiste en escribir un pequeño programa (un "script") que haga exactamente eso por ti en cuestión de segundos.

## 1.2 Usos comunes del Web Scraping
En el ámbito académico y profesional, el scraping se usa para:

- **Investigación de mercado**: Comparar precios y productos de diferentes competidores automáticamente.

- **Ciencia de Datos**: Recolectar grandes volúmenes de texto para entrenar modelos de Inteligencia Artificial.

- **Monitoreo de noticias**: Seguir menciones de un tema específico en portales de noticias o redes sociales.

- **Análisis de sentimientos**: Extraer reseñas de usuarios para saber si un producto o política pública gusta o no.

- **Investigación Académica**: Recopilar publicaciones, citas o datos estadísticos de portales gubernamentales que no tienen un botón de "descargar".

## 1.3 Diferencias entre Web Scraping y Web Crawling

<img src="assets/scrapping_vs_crawwling.png" width="40%"/>

A menudo estos términos se usan como sinónimos, pero en el mundo técnico tienen objetivos distintos:

- **Web Scraping (Extracción)**: Se enfoca en obtener datos específicos de páginas determinadas. Por ejemplo: "Extraer el nombre y el precio de todos los productos en esta página de Amazon".

- **Web Crawling (Navegación)**: Es el proceso de descubrir nuevas páginas. Es lo que hacen los motores de búsqueda como Google o Bing. El "crawler" entra a una web, sigue todos los enlaces que encuentra para saltar a otras páginas y así "mapear" el internet.

## 2.0.1 Anatomía de una URL: Desglosando el mapa
Vamos a desarmar una URL típica para ver qué partes nos interesan para el scraping:

<img src="assets\estructura-url.png" width="60%"/>

### 1. Protocolo `(http:// o https://)`
Es el "idioma" en el que se comunican tu navegador y el servidor. El https (la "s" es de secure) es el estándar actual. En Python, la librería que usaremos (requests) necesita saber esto para establecer la conexión.

Como dato, si una página no tiene https, es posible que el servidor no permita la conexión o que la información viaje sin cifrar, lo que puede ser un riesgo de seguridad.
Y es justo lo que pasó con la página de datos abiertos del gobierno del Ecuador. Al no tener https, el servidor no permitia acceso y no pudimos hacer scrapping por 2 meses :)

### 2. Dominio ``(www.scraperapi.com)``
Es el nombre del servidor donde vive la página. Es como la dirección de un edificio.

### 3. Ruta o Path ``(/productos/libros)``
Indica la ubicación específica dentro del servidor. Es como el número de oficina o departamento dentro del edificio.

Ejemplo: tienda.com/ciencia nos llevará a una sección distinta que tienda.com/historia.

### 4. Parámetros de consulta o Query Strings ``(?categoria=biologia&page=2)``
¡Aquí es donde sucede la magia para el scraping! Los parámetros empiezan después del signo ? y se separan por el símbolo &.

Son pares de llave=valor.

**¿Por qué son importantes?** Porque nos permiten modificar lo que la página muestra sin navegar manualmente. Si vemos que al cambiar page=2 por page=3 la web nos muestra más resultados, ¡podemos hacer un bucle en Python para extraer 100 páginas de datos en un segundo!

### 5. Fragmento o Ancla ``(#seccion-referencias)``
Aparece después de un #. Indica un lugar específico dentro de la misma página.

Nota técnica: Generalmente, el servidor ignora lo que va después del #, por lo que para el scraping básico solemos ignorarlo, ya que el contenido suele ser el mismo.


## 2.1 Explicación de HTML: El esqueleto de la web
HTML (*HyperText Markup Language*) no es un lenguaje de programación, sino un lenguaje de marcado. Sirve para decirle al navegador: "esto es un título", "esto es un párrafo" o "esto es una imagen".

Para un scraper, el HTML es como una estantería organizada. Si sabemos en qué "caja" está la información, podemos pedirle a Python que la tome por nosotros.


### 2.1.1 Etiquetas HTML (Tags)
Las etiquetas son los ladrillos de la web. Casi siempre vienen en parejas: una de apertura y una de cierre.

Aquí están las etiquetas que más te vas a encontrar:

- ``<!DOCTYPE html>``: Define que el documento es HTML5.

- ``<html>``: Es el elemento raíz que contiene todo el código de la página.

- ``<head>``: Contiene la "meta-información" (datos invisibles para el usuario, como el título de la pestaña).

- ``<title>`` : Especifica el título que aparece en la pestaña del navegador.

- ``<body>``: Contiene todo el contenido visible de la página (lo que realmente queremos scrapear).

- ``<div>``: Define una división o sección. Es la etiqueta más común y funciona como un contenedor para organizar otros elementos.

- ``<h1>`` a ``<h6>``: Definen encabezados o títulos, siendo ``<h1>`` el más importante (título principal).

- ``<p>``: Define un párrafo de texto.

- ``<a>``: Define un hipervínculo. Se usa junto al atributo href, que contiene la URL a la que apunta el enlace.
    - Ejemplo: <a href="https://google.com">Ir a Google</a>

- `` <ul>``: Define una lista desordenada (con viñetas o puntos).

- ``<li>``: Define cada uno de los elementos de una lista (vive dentro de un ``<ul>`` o ``<ol>``).

- ``<table>``: Define una tabla de datos.

- ``<tr>``: Define una fila dentro de una tabla (table row).

- ``<td>``: Define una celda de datos dentro de la fila (table data). Es aquí donde suele estar el valor que buscamos.

Atributos: Las etiquetas pueden tener "apellidos" llamados atributos. Los más importantes para nosotros son class e id, porque funcionan como etiquetas de nombre que nos permiten identificar un elemento específico entre miles.

### 2.1.2 El DOM (Document Object Model) 🌳
El DOM es la estructura jerárquica de una página web. Imagina que el HTML es el código escrito, pero el DOM es el "árbol" vivo que el navegador construye en su memoria.

¿Por qué nos importa para el Scraping?
Porque gracias al DOM, Python puede "navegar" por la página. Podemos decirle: "Ve al cuerpo (body), busca la segunda división (div) y dame el texto del tercer párrafo (p)". Es una estructura de padres e hijos.

## 2.2 Explicación de CSS:
El estilo que nos ayuda a filtrarSi el HTML es la estructura, el CSS (Cascading Style Sheets) es lo que le da el diseño (colores, tamaños, posiciones).

Para nosotros, el CSS no es importante por la estética, sino por los selectores. Los diseñadores web usan "etiquetas de nombre" para aplicar estilos, y nosotros usaremos esas mismas etiquetas para identificar los datos.
## 2.2.1 Clases e IDs: 
Nuestros mejores aliadosCuando hay 50 etiquetas <div> en una página, ¿cómo sabemos cuál es la que tiene el precio del producto? Usamos selectores:

- ID (``#``): Es un identificador único. Solo debería haber uno por página. Es como el número de cédula o DNI de un elemento.
    - Ejemplo: ``<h1 id="titulo-principal">`` $\rightarrow$ Selector CSS: ``#titulo-principal``
- Clases (``.``): Se usan para identificar grupos de elementos. Muchos elementos pueden compartir la misma clase.
    - Ejemplo: ``<span class="precio">`` $\rightarrow$ Selector CSS: ``.precio``

**Regla de oro**: Si encuentras un ID, ya ganaste; es la forma más fácil de extraer un dato. Si no hay ID, buscamos las Clases.

## 3.2 Requests: Nuestro mensajero 📡
Para obtener los datos de una web, primero debemos "traerla" a Python. La librería requests es el estándar de oro para hacer esto.

**Instalación y uso básico**

En Google Colab, requests ya viene preinstalado, igual vamos a instalarlo para que aprendan.

Para llamar a librerias utilizamos la sentencia ``import``

````python
# Asi lo instalan librerias en un notebook
%pip install requests
# Asi llaman a las librerias
import requests
# Definimos la URL de nuestro caso de estudio
URL = "https://realpython.github.io/fake-jobs/"

# Realizamos la petición GET
respuesta = requests.get(URL)

# Verificamos qué nos respondió el servidor
print(f"Código de estado: {respuesta.status_code}") # 200 significa éxito
print("-" * 30)
# Mostramos los primeros 250 caracteres del HTML crudo
print(respuesta.text[:250])
````

Dato clave: Lo que ves arriba es el HTML "crudo". Es un texto gigante y difícil de leer para un humano, pero es donde están nuestros datos.

### 3.2.1 Entendiendo los Status Codes (Códigos de Estado) 🚦

Cada vez que enviamos al "mensajero" (`requests.get()`), el servidor nos devuelve un sobre con un número de tres dígitos. Este número es el **Status Code** y nos dice qué pasó con nuestra petición.

Es fundamental revisarlo porque no queremos intentar extraer datos de una página que no cargó o que nos bloqueó.

#### Los códigos más comunes en Scraping:

| Código | Significado | ¿Qué significa para nosotros? |
| :--- | :--- | :--- |
| **200** | **OK** | ¡Éxito! Tenemos el HTML listo para trabajar. |
| **403** | **Forbidden** | El acceso está prohibido. El sitio detectó que eres un bot y te bloqueó. |
| **404** | **Not Found** | La página no existe. Revisa si escribiste bien la URL. |
| **500** | **Internal Server Error** | La página está caída o tiene problemas técnicos propios. |

## 3.3 Regex: El método rudimentario 🔍

Antes de que existieran librerías inteligentes, los programadores usaban **Regex** (Expresiones Regulares). Una expresión regular es una secuencia de caracteres que forma un **patrón de búsqueda**.

Imagina que el HTML de la página es un bosque gigante y tú estás buscando una flor específica. Regex es como intentar encontrar esa flor describiendo exactamente cómo son sus pétalos y su tallo, carácter por carácter.

### ¿Por qué decimos que es rudimentario para el Scraping?

1. **Sensibilidad extrema:** El HTML no es texto plano, es código. Si el desarrollador de la página web agrega un espacio extra, cambia una comilla doble (`"`) por una simple (`'`) o salta de línea, tu patrón de Regex se romperá y no encontrará nada.
2. **No entiende la jerarquía:** Como vimos antes, la web tiene una estructura de "padres e hijos" (DOM). Regex no entiende eso; solo ve una cadena larga de texto. No sabe si un título está dentro de un encabezado o de un pie de página.
3. **Mantenimiento imposible:** Las expresiones regulares suelen ser difíciles de leer y aún más difíciles de corregir.

### Ejemplo de "Fuerza Bruta" con Regex

Supongamos que queremos extraer el título del primer trabajo.

 Sabemos que en el HTML está escrito así:
`<h2 class="title is-5">Python Developer</h2>`

Para extraerlo con Regex, tendríamos que escribir algo como esto:


````python
import re

# El contenido de la página que descargamos con requests
html_crudo = respuesta.text

# Intentamos definir el patrón "manual"
# Este patrón dice: "Busca lo que esté después de esa clase de h2 y antes de que cierre el h2"
patron = r'<h2 class="title is-5">(.+?)</h2>'

resultados = re.findall(patron, html_crudo)

print(f"Se encontraron {len(resultados)} títulos.")
print(f"Ejemplo de los títulos: {resultados[0:5]}")
````

> **Reflexión para la clase:** ¿Qué pasaría si mañana el dueño de la página decide que la clase ya no es `is-5` sino `is-6`?

> Exacto: **nuestro código dejaría de funcionar por completo.**

Esta sección es donde los chicos verán cómo pasamos de "pelear" con el texto a "conversar" con la estructura de la página. **Beautiful Soup** es la librería que convierte el caos del HTML en algo ordenado y fácil de usar.

## 3.4 Beautiful Soup: El estándar para investigación 🥣

**Beautiful Soup** es una librería de Python diseñada para extraer datos de archivos HTML y XML. A diferencia de Regex, esta herramienta "entiende" la estructura de la página (el DOM que vimos antes) y nos permite navegar por ella de forma lógica.

Imagina que, en lugar de buscar una cadena de texto, ahora puedes decirle a Python: *"Tráeme todos los títulos de nivel 2 que tengan la clase 'title'"*.

### 3.4.1 Instalación y configuración

En un entorno local, la instalarías con `pip install beautifulsoup4`. Como estamos en **Google Colab**, la librería ya viene instalada, pero necesitamos importarla en nuestro código.

También necesitamos un "parser" (un analizador), que es el motor que procesa el texto. Usaremos el que viene por defecto en Python: `html.parser`.

### 3.4.2 Uso básico: "Cocinando" la sopa

El primer paso siempre es tomar el texto crudo de la página (lo que nos trajo `requests`) y pasárselo a Beautiful Soup para crear un objeto llamado **`soup`**.


````python
%pip intall bs4
from bs4 import BeautifulSoup

# 1. Tomamos el contenido de la respuesta que obtuvimos con requests
# 2. "Cocinamos" el HTML para crear nuestro objeto 'soup'
soup = BeautifulSoup(respuesta.text, 'html.parser')

# Ahora 'soup' es un objeto que representa toda la estructura de la página
print("¡Objeto BeautifulSoup creado con éxito!")
````


### 3.4.3 Buscando elementos: `find()` y `find_all()`

Estas son las dos funciones que más usarán en su carrera como investigadores:

* **`find()`**: Te devuelve el **primer** elemento que coincida con lo que buscas. Es ideal para cosas únicas como el título principal o un ID específico.
* **`find_all()`**: Te devuelve una **lista** con todos los elementos que coincidan. Es perfecto para extraer listas de precios, nombres o, en nuestro caso, ofertas de trabajo.

**Ejemplo rápido:**


````python
# Buscamos el primer título h1 de la página
titulo_principal = soup.find("h1")
print(f"Título de la página: {titulo_principal.text.strip()}")

# Buscamos TODOS los títulos de las ofertas de trabajo
# En esta página, los títulos son h2 y tienen la clase 'title'
titulos_ofertas = soup.find_all("h2", class_="title")

print(f"Hemos encontrado {len(titulos_ofertas)} ofertas de trabajo.")
print(f"Ejemplo de los títulos: {titulos_ofertas[0:5]}")
````


**¿Por qué `class_` con un guion bajo?**

Notarás que en el código escribimos `class_="title"`. 

Esto es porque en Python la palabra `class` está reservada para crear clases de programación. Para evitar confusiones, Beautiful Soup usa `class_`.



# 4. Extrayendo información específica 💎

Ya tenemos nuestra "sopa" (el objeto `soup`) lista. Ahora vamos a aprender a extraer exactamente lo que nos interesa de cada etiqueta.

## 4.1 Extraer el texto limpio con `.text` y `.strip()`

Cuando usamos `find()` o `find_all()`, Beautiful Soup nos devuelve la etiqueta completa, por ejemplo: `<h2 class="title is-5"> Python Developer </h2>`. Pero para nuestra base de datos, solo queremos el nombre del puesto.

* **`.text`**: Extrae solo el contenido textual dentro de la etiqueta.
* **`.strip()`**: Es una función de Python que elimina los espacios en blanco innecesarios al principio y al final (muy común en el código HTML).


````python
# Tomamos la primera oferta de trabajo
primera_oferta = soup.find("h2", class_="title")

# Extraemos solo el texto
texto_sucio = primera_oferta
texto_limpio = primera_oferta.text.strip()

print(f"Texto crudo: '{texto_sucio}'")
print(f"Texto limpio: '{texto_limpio}'")
````


## 4.2 Extraer atributos con corchetes `[]`

A veces el dato no está "escrito" en la página, sino que está escondido dentro de un atributo de la etiqueta. El caso más común es el enlace de una vacante, que vive dentro del atributo **`href`** de una etiqueta `<a>`.

Para obtenerlo, tratamos a la etiqueta como si fuera un **diccionario** de Python:


````python
# Buscamos el primer enlace de la página
enlace = soup.find("a")

# Extraemos el valor del atributo 'href'
url_destino = enlace['href']

print(f"El enlace con el texto '{enlace.text}' apunta a: {url_destino}")
````


## 4.3 Unificando todo: El bucle de extracción

Ahora que sabemos extraer el texto y los atributos, podemos recorrer todas las ofertas para crear una lista organizada.


````python
# 1. Buscamos todas las cajas que contienen las ofertas
# En esta web, cada trabajo está en un <div> con clase 'card-content'
ofertas = soup.find_all("div", class_="card-content")

print(f"Procesando {len(ofertas)} ofertas...\n")

for oferta in ofertas:
    # Dentro de cada caja, buscamos el título y la empresa
    titulo = oferta.find("h2", class_="title").text.strip()
    empresa = oferta.find("h3", class_="company").text.strip()
    # Buscamos el link (está en el segundo enlace de la caja)
    link = oferta.find_all("a")[1]["href"]
    print(f"Puesto: {titulo}")
    print(f"Empresa: {empresa}")
    print(f"Link: {link}")
    print("-" * 20)

````

## 5.2 ¿Cómo se hace realmente un scraper? (Automatización)

Hasta ahora hemos corrido bloques de código manualmente. Pero en la vida real, un **scraper profesional** no vive en un notebook de prueba; vive en un archivo de script (`.py`) que se puede programar para ejecutarse solo.

**Los componentes de un scraper real son:**

1. **El Loop (Bucle):** Si tienes que scrapear 100 páginas, no escribes el código 100 veces. Usas un bucle `for` que cambie la URL automáticamente.
2. **Manejo de Errores:** ¿Qué pasa si el internet se corta en la página 50? Un buen scraper sabe esperar y reintentar.
3. **Respeto al Servidor (Etiqueta):** Usamos pequeñas pausas (`time.sleep`) para no saturar el servidor de la página web con miles de peticiones por segundo.


## 5.3 ¡Manos a la obra! Segundo ejemplo: Librería Online 📚

Ahora vamos a aplicar todo lo aprendido a un sitio diferente: [books.toscrape.com](http://books.toscrape.com/).

<img src="assets\books_to_scrape.png" width="50%"/>

**El Reto:** Queremos extraer el **título**, el **precio**, el **número de estrelllas**, la **disponibilidad** de todos los libros que aparecen en la página principal hasta la página 5.

### El Análisis Preliminar (Inspección)

Si haces clic derecho en un libro e inspeccionas, notarás que:

1. Cada libro está dentro de una etiqueta `<article>` con la clase `product_pod`.
2. El título está dentro de un enlace `<a>` que vive dentro de un `<h3>`.
3. El precio está en un `<p>` con la clase `price_color`.
4. La disponibilidad está en un `<p>` con la clase `instock availability`.
5. *Pregunta: ¿En qué clase está el número de estrellas?*


### El Scraper Completo

Aquí tienes el código unificado. Fíjate cómo seguimos el flujo: Petición $\rightarrow$ Sopa $\rightarrow$ Bucle $\rightarrow$ Extracción $\rightarrow$ Guardado.


````python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

# 1. Configuración
URL = "http://books.toscrape.com/"
print(f"Iniciando el scraping en: {URL}...")
# Lista para almacenar todos los libros de todas las páginas
todos_los_libros = []


for pagina in range(1, 6):
    # Construimos la URL dinámica para cada página
    url = f"http://books.toscrape.com/catalogue/page-{pagina}.html"
    print(f"🔎 Analizando página {pagina}...")
    
    respuesta = requests.get(url)
    if respuesta.status_code == 200:
        soup = BeautifulSoup(respuesta.text, 'html.parser')
        libros = soup.find_all('article', class_='product_pod')
        
        for libro in libros:
            # 1. Título completo (del atributo 'title')
            titulo = libro.h3.a['title']
            
            # 2. Precio
            precio = libro.find('p', class_='price_color').text.strip()


            # 3. Estrellas (están en la clase de la etiqueta <p>)
            # La clase es algo como ['star-rating Three']
            # al escoger 1 es el segundo elemento
            clases_estrellas = libro.find('p', class_='star-rating')['class']
            estrellas = clases_estrellas[1] # Tomamos la segunda palabra (Three, Two, etc.)
            
            # 4. Disponibilidad
            stock = libro.find('p', class_='instock availability').text.strip()
            
            # Guardamos la info del libro
            todos_los_libros.append({
                "Título": titulo,
                "Precio": precio,
                "Estrellas": estrellas,
                "Disponibilidad": stock,
                "Página": pagina
            })
        
        # Buena práctica: esperar un segundo entre páginas para no saturar el servidor
        time.sleep(1)
    else:
        print(f"⚠️ Error en página {pagina}: Código {respuesta.status_code}")

# Convertir a DataFrame y guardar
df_final = pd.DataFrame(todos_los_libros)
df_final.to_csv("libros_completos.csv", index=False)

print(f"\n✅ ¡Proceso terminado! Se extrajeron {len(df_final)} libros en total.")
print("El archivo 'libros_completos.csv' ya está listo.")
````


### ¿Por qué este scraper es mejor?

* **Es modular:** Puedes entender cada paso fácilmente.
* **Usa Selectores Precisos:** Al usar `class_='product_pod'`, nos aseguramos de no traer información basura de los menús laterales o del pie de página.
* **Usa Atributos:** En el caso del título, usamos `libro.h3.a['title']` porque a veces el texto visible está cortado (ej. "A Light in the..."), pero el atributo `title` siempre tiene el nombre completo.

* **La solución:** Necesitas un "navegador real" controlado por código. Herramientas como **Selenium** o **Playwright** abren una ventana de Chrome o Firefox, esperan a que el JavaScript se ejecute y los datos aparezcan, y recién ahí extraen la información.

# 7. Consideraciones legales y éticas del web scraping ⚖️

El hecho de que los datos sean visibles en un navegador no significa que tengamos permiso absoluto para tomarlos de cualquier manera. Como investigadores, debemos actuar bajo principios éticos para no perjudicar a los dueños de los sitios ni incurrir en faltas legales.

## 7.1 Legalidad del web scraping

El web scraping **no es ilegal en sí mismo**, pero su uso indebido puede traer consecuencias. Es similar a tomar fotos en la calle: puedes hacerlo, pero si entras sin permiso a una propiedad privada o usas las fotos de alguien más para lucrar sin autorización, tienes un problema.

**Es ilegal o problemático si:**

* Violas los **Términos de Servicio (ToS)** que aceptaste al crear una cuenta.
* Accedes a información protegida por **derechos de autor** para redistribuirla como propia.
* Recolectas **datos personales** sensibles (nombres, correos, teléfonos) sin consentimiento, violando leyes de privacidad (como el GDPR).
* Tu scraper hace tantas peticiones que **tira el servidor** (esto se considera un ataque de denegación de servicio o DoS).


## 7.2 Robots.txt: El manual de convivencia del sitio

El archivo `robots.txt` es un protocolo de exclusión. Es un archivo de texto que los dueños de las webs ponen en su servidor para decirle a los bots (como Google o tu script de Python) qué partes de la página pueden visitar y cuáles están prohibidas.

### 7.2.1 Ejemplos de robots.txt

Puedes ver el archivo de cualquier página agregando `/robots.txt` al final de su URL. Veamos algunos ejemplos:




````python
urls = [
    'https://www.nesta.org.uk/robots.txt',
    'https://www.boxofficemojo.com/robots.txt',
    'https://www.howtogeek.com/robots.txt'
]

for u in urls:
    print(f"--- Robots.txt de: {u} ---")
    try:
        print(requests.get(u).text[:300]) # Mostramos solo el inicio
    except:
        print("No se pudo acceder.")
    print("\n")
````


**¿Qué buscar en un robots.txt?**

* **`User-agent: *`**: Instrucciones para todos los bots.
* **`Allow: /`**: Indica qué rutas sí se pueden scrapear.
* **`Disallow: /admin/`**: Rutas prohibidas para los scrapers.


## 7.3 User-Agents: "Identifícate al entrar"

Cuando haces una petición con `requests`, por defecto Python le dice al servidor: *"Hola, soy un script de Python"*. Muchos servidores bloquean esto de inmediato.

El **User-Agent** es una cadena de texto que le dice al servidor qué navegador y sistema operativo estás usando.

> **Práctica Ética:** Identifícate. En proyectos de investigación serios, es recomendable incluir tu nombre o correo en el User-Agent para que el administrador del sitio pueda contactarte si tu scraper está causando problemas, en lugar de simplemente bloquear tu IP.

```python
# Ejemplo de cómo enviar un User-Agent personalizado
headers = {
    'User-Agent': 'Investigador Academico (contacto: mi-correo@universidad.edu)',
    'Accept-Language': 'es-ES,es;q=0.9'
}
respuesta = requests.get("https://ejemplo.com", headers=headers)

```


## 7.4 Reglas del Scraper Ético 📜

Para que tu investigación sea impecable, sigue estas reglas:

1. **Usa la API primero:** Si el sitio tiene una API oficial, úsala. Es más estable y legal.
2. **No copies, crea valor:** No dupliques el sitio web. Extrae datos para generar nuevos conocimientos o análisis.
3. **Sé amable con los sitios pequeños:** No pidas 100 páginas por segundo. Usa `time.sleep()` para espaciar tus peticiones.
4. **Scrapea en horas bajas:** Si el sitio es local, corre tu script de madrugada cuando hay menos tráfico de usuarios reales.
5. **No mientas:** No intentes pasar datos scrapeados como si fueran una encuesta propia. Siempre da crédito a la fuente.

# 8. ¿Qué otras herramientas existen para hacer Web Scraping? 🛠️

A medida que tus proyectos de investigación crezcan en volumen o dificultad, notarás que `BeautifulSoup` se queda corto. Aquí te presento las herramientas "pesadas" que usan los profesionales.

## 8.1 Scrapy: El "Framework" Industrial
<img src="assets\scrapy.png" width=70%>

Si `BeautifulSoup` es un destornillador, **Scrapy** es una fábrica completa.Su funcionamiento no es de una librería unicamente, en realidad es un marco de trabajo (framework) diseñado para el scraping a gran escala.

* **Velocidad:** Es extremadamente rápido porque puede procesar muchas peticiones al mismo tiempo (asíncrono).
* **Gestión Integrada:** Ya viene preparado para manejar cookies, sesiones, User-Agents y guardar datos en bases de datos automáticamente.

<img src="assets\selenium.png" width=70%>

Como mencionamos antes, **Selenium** no fue creado para scraping, sino para automatizar pruebas en sitios web. Sin embargo, es el "arma secreta" para las páginas más difíciles.

* **Potencia:** Al ser de los mismos creadores de Chrome, es increíblemente rápido y eficiente para interactuar con sitios web modernos. Es muy popular entre desarrolladores de software, aunque para investigadores, Python sigue siendo el rey por su facilidad.

# 9. La manera más fácil de obtener información: APIs 🔌

Antes de lanzar tus "arañas" a recorrer el código HTML, hay un secreto que todo investigador debe conocer: a veces, los sitios web nos dan los datos voluntariamente y de forma ordenada.

## 9.1 ¿Qué es una API?

**API** significa *Application Programming Interface* (Interfaz de Programación de Aplicaciones). Imagina que vas a un restaurante:

* El **HTML** es como entrar a la cocina a escondidas para ver qué ingredientes hay (Scraping).
* La **API** es el **menú** que te entrega el mesero: tú pides exactamente lo que quieres y la cocina te lo entrega en una bandeja, limpio y listo para comer.

En términos técnicos, una API es un canal oficial que permite que tu código de Python le pida datos directamente a la base de datos del servidor, sin tener que pasar por el diseño visual de la página.

## 9.2 ¿Por qué es mejor usar una API que hacer Web Scraping?

Si una página tiene API, **úsala siempre**. Aquí las razones:

1. **Legalidad y Ética:** Al usar la API, estás siguiendo las reglas del sitio. No hay riesgo de demandas ni bloqueos por comportamiento "sospechoso".
2. **Estabilidad:** El HTML cambia constantemente (un rediseño del sitio rompería tu scraper). La API es una estructura fija que rara vez cambia.
3. **Limpieza:** No tienes que limpiar etiquetas `<div>` o `<span>`. La API te entrega los datos en formato **JSON**, que es básicamente un diccionario de Python listo para usar.
4. **Rapidez:** No descargas imágenes, estilos CSS ni anuncios; solo descargas los datos puros.

## 9.3 Ejemplos de APIs famosas

Muchos de los datos que los investigadores necesitan ya están disponibles a través de estos canales:

* **API de Reddit:** Muy abierta y amigable para investigadores que analizan sentimientos o tendencias sociales.
* **API de Google Maps:** Para obtener coordenadas, distancias o información de lugares.
* **API de Twitter (X):** Históricamente la más usada para análisis político y social (aunque ahora tiene más restricciones).
* **API de Spotify:** Ideal para analizar géneros musicales y popularidad de artistas.
* **APIs de Gobiernos (Datos Abiertos):** Muchos países tienen APIs para consultar el clima, el presupuesto público o estadísticas demográficas.


> **Consejo Pro:** Busca en Google el nombre del sitio seguido de la palabra "API" (ej: *"YouTube API"*). Si existe, te ahorrarás el 90% del trabajo.


## 9.4 El costo de la comodidad: ¿Son gratis las APIs? 💸

Aunque muchas APIs ofrecen un "nivel gratuito" para estudiantes o investigadores, no siempre son gratis. A diferencia del scraping (donde el costo es principalmente tu tiempo y código), las APIs suelen funcionar bajo estos modelos:

* **Freemium:** Te permiten hacer un número limitado de peticiones al mes (ej. 1,000 consultas gratis) y te cobran si te excedes.
* **Suscripción:** Pagas una cuota mensual por acceso total a los datos.
* **Pago por uso (Pay-as-you-go):** Pagas exactamente por cada dato que pides (común en servicios de mapas o traducción).

> **Nota para el investigador:** Siempre revisa la página de "Pricing" de la API antes de empezar, para evitar sorpresas en tu tarjeta de crédito.

## 9.5 APIs más allá de los datos: El caso de la IA 🤖

Hoy en día, las APIs no solo sirven para traer "listas de cosas". Son la forma en que conectamos nuestros programas con cerebros artificiales.

El ejemplo más famoso actualmente es la **API de OpenAI** (la empresa detrás de ChatGPT).

* **¿Cómo funciona?** En lugar de entrar a la web de ChatGPT y escribir a mano, tu script de Python le envía una pregunta a través de la API y recibe la respuesta de la IA directamente para procesarla.
* **Uso en investigación:** Puedes usar esta API para que una IA lea miles de comentarios que scrapeaste y los clasifique automáticamente en "positivos" o "negativos". O cómo hicimos en el Datahub, que la utilizamos para extraer campos de las sentencias judiciales para analizar las secciones que no podemos hacer con regex

**Otros ejemplos de APIs funcionales:**

* **Stripe API:** Para procesar pagos en una web.
* **Twilio API:** Para que tu código envíe mensajes SMS o WhatsApp automáticamente.
* **OpenWeather API:** Para obtener el clima exacto de cualquier ciudad en tiempo real.
